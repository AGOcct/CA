# obtaining County name from coordinates using Nominatim API

```python
#pip install geopy
```

```python
t=op.iloc[:5]
```

```python
t
```

```python
from geopy.geocoders import Nominatim

# Initialize Nominatim API
geolocator = Nominatim(user_agent="MyApp")

# function to reverse geocode a row of coordinates and extract county 
def reverse_geocode(row):
    location = geolocator.reverse((row['latitude'], row['longitude']))
    address = location.raw['address']
    county = address.get('county','')
    return f"{county}"



# reverse geocoding function to each row
op['location'] = t.apply(reverse_geocode, axis=1)

op
```

## exctract county

```python
#creating list of existing county
county_list = ['Dublin', 'Laois', 'Meath', 'Kilkenny', 'Limerick', 'Carlow', 'Cork', 'Clare', 'Sligo', 'Cavan', 'Tipperary',
               'Wicklow', 'Roscommon', 'Wexford', 'Mayo', 'Donegal', 'Longford', 'Galway', 'Offaly', 'Kildare',
               'Waterford', 'Louth', 'Kerry', 'Westmeath', 'Monaghan', 'Leitrim']

def extr_name(link):
    if isinstance(link, str):
        for location in county_list:
            if location.lower() in link.lower():
                return location
    return None

#create a new 'Extracted_County'
op['Extracted_County'] = op['preprocessed_url'].apply(extr_name)

#Replace NaN values in the 'County' column with the extracted county name
op['location'] = op['location'].fillna(op['Extracted_County'])

# Drop the 'Extracted_County' 
op = op.drop('Extracted_County', axis=1)
```
